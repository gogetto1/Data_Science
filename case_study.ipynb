#Churn Modelling

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://ml-repository-krakers.s3-eu-west-1.amazonaws.com/kaggle+/churn_modelling/Telco-Customer-Churn.csv'

df = pd.read_csv(url)
df.head(5)
```

```
df.info()
```

```
TotalChargesMedian = df[df.TotalCharges != ' ']['TotalCharges'].astype('float32').median()
```

```
df['TotalCharges'][df['TotalCharges'] == ' '] = TotalChargesMedian
df['TotalCharges'] = df['TotalCharges'].astype('float32')
```

```
df.info()
```

```
df.isnull().sum().sum()
```

```
categorical = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
               'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'Contract', 'StreamingMovies',
               'PaperlessBilling', 'PaymentMethod', 'Churn']
 
numerical = ['tenure', 'MonthlyCharges', 'TotalCharges']

for col in categorical:
  df[col] = pd.Categorical(df[col])

for col in numerical:
  df[col] = df[col].astype('float32')

df.info()
```

```
df = df.drop(columns=['customerID'])
df.info()
```

```
df.describe()
```

```
df.describe(include=['category'])
```

```
plt.figure(figsize=(8, 6))
sns.set()
df.Churn.value_counts().plot(kind='pie',
                             fontsize=16,
                             colors=['#3DDC97', '#FF495C'],
                             explode=[0.1, 0],
                             shadow=True,
                             autopct='%1.1f%%'
                             )
plt.legend()
plt.ylabel('')
plt.title('Rozk≈Çad Churn')
print(df.Churn.value_counts())
```

```
plt.figure(figsize=(8, 6))
sns.set()
df.gender.value_counts().plot(kind='pie',
                             fontsize=16,
                             colors=['#3DDC97', '#FF495C'],
                             explode=[0.1, 0],
                             shadow=True,
                             autopct='%1.1f%%',
                              labels=['', '']
                             )
plt.legend(['Male', 'Female'])
plt.ylabel('')
plt.title('Rozk≈Çad p≈Çci')
print(df.Churn.value_counts())
```

```
plt.figure(figsize=(8, 6))
sns.set()
df.SeniorCitizen.value_counts().plot(kind='pie',
                             fontsize=16,
                             colors=['#3DDC97', '#FF495C'],
                             explode=[0.1, 0],
                             shadow=True,
                             autopct='%1.1f%%',
                             )
plt.legend()
plt.ylabel('')
plt.title('Rozk≈Çad SeniorCitizen')
print(df.Churn.value_counts())
```
